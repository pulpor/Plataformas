```python
import pandas as pd

#1 - importar a base de dados

tabela = pd.read_csv("telecom_users.csv")  #não precisa do caminho, pois está na mesma pasta

#2 - visualizar e entender as informações disponíveis 


#drop tem 2 parametros ("nome", axis)

#axis é eixo.   se for = 0 -> linha  ,  se for = 1 -> coluna

tabela = tabela.drop("Unnamed: 0", axis=1)   # 'drop' é 'largar'
display (tabela)

```

```python
#3 - tratamento de dados 

# tratamento 1 ) resolver valores que estão sendo reconhecidos de forma errada

#além de transformar, você deve tratar os erros (algum burrão escreveu em vez de tacar numeros)
#errors = "coerce" = transforma os erros em vazios  ('coerce' significa 'forçar')

tabela["TotalGasto"] = pd.to_numeric(tabela["TotalGasto"], errors="coerce")


# tratamento 2 ) resolver valores vazios


#Ao usar drop somente, descarta as colunas/linhas que você definir
#Ao usar dropna, remove todas as entradas com valores NaN [not a number, que n tem numeros] (ou null em geral)

#axis ( 1 = coluna e 0 = linha)

#colunas que tem TODOS os valorez vazios - usar ALL         (deletando colunas vazias)
tabela = tabela.dropna(how="all", axis=1)

#colunas que tem PELO MENOS 1 os valor vazio - usar ANY     (deletando linhas vazias)
tabela = tabela.dropna(how="any", axis=0)
#NaN = not a number

print(tabela.info())
```

```python
display(tabela)
```

```python
#4 - analise inicial

print(tabela['Churn'].value_counts())                   #conta quantos churn tem
print(tabela["Churn"].value_counts(normalize=True))     #conta em percentual 


#5 - analise detalhada, descobrir as causas do cancelamento
```

```python
import plotly.express as px 

#criar grafico - histograma - [tabela, quem sao os valores do eixo x, diferenciação de cor com outro parametro]

coluna = "Aposentado"
comparacao = "Churn"
grafico = px.histogram(tabela, x=coluna, color=comparacao)

#px.barplot  -> coluna
#px.piechar  -> pizza
#px.pie      -> linha

#exibir o grafico
grafico.show()
```
