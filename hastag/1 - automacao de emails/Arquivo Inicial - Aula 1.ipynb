```python
import pyautogui
import pyperclip #soluciona a limitação (de caractere especial) do pyautogui
import time

pyautogui.PAUSE = 1  #pausa a cada 1 segundo.

# pyautogui.click -> clicar
# pyautogui.write -> escrever
# pyautogui.press -> pressionar
# pyautogui.hotkey -> atalho

pyautogui.press("win")
pyautogui.write("chrome")

pyautogui.press("enter")

pyautogui.hotkey('ctrl','t') #abrir uma aba nova no navegador 
pyperclip.copy("https://drive.google.com/drive/folders/14oLE59U1RqyRqlBbKpsyymW-mitvbtoh")
pyautogui.hotkey('ctrl','v')
pyautogui.press("enter")

time.sleep(5) #espera 5 segundos 
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas as pd  # as 'pd' é como se fosse um apelido para a biblioteca pandas

tabela = pd.read_excel(r"D:\teste de down\Vendas - Dez.xlsx")
# r é de 'raw' significa arquivo *cru* (pra não utilizar os comandos '\t' ou '\n' ou '\v')

display(tabela) #display deixa a tabela mais bonitinha que o print

#pegar uma coluna utiliza-se colchetes
faturamento = tabela ["Valor Final"].sum()  #sum é soma 
quantidade = tabela ["Quantidade"].sum()

print(faturamento)
print(quantidade)

pyautogui.PAUSE = 1

# passo 5: abrir o email
pyautogui.press("win")
pyautogui.write("chrome")
pyautogui.press("enter")
pyautogui.write("www.gmail.com")
pyautogui.press("enter")

time.sleep(2)
pyautogui.click(x=115, y=248, clicks=1)

#passo 6: abrir aba para email   #pegar a posição do mouse pelo .position() com time sleep de 5s

time.sleep(5)
pyautogui.write("pulppor@hotmail.com")
pyautogui.press("tab")     #mandar mais de uma pessoa, só dar tab.

pyautogui.write("pultirinhas@gmail.com")
pyautogui.press("tab")

pyautogui.write("leonardo.pulpor@hotmail.com")
pyautogui.press("tab")
pyautogui.press("tab")      #dois pulos muda p linha de baixo (assunto)

#assunto do email
pyperclip.copy("Textão Maluco!")   #caractere especial precisa ser copiado e colado
pyautogui.hotkey("ctrl","v")
pyautogui.press("tab")


time.sleep(2)
#corpo do email
#3 aspas para escrever dando várias quebras de linhas (legal né rs)
texto = f""" Prezados, bom dia

o faturamento de ontem foi de: R${faturamento:,.2f}   
a quantidade de produtos foi de: {quantidade:,}

Abraços, Pulpor."""

# dois pontos é p formatar (vírgula p separar e f é de float)
pyperclip.copy(texto)   #caractere especial precisa ser copiado e colado
pyautogui.hotkey("ctrl","v")

time.sleep(3)
#9 - enviar e-mail
# pyautogui.click(x=508, y=1013)   -> clica em enviar pelo mouse 

pyautogui.hotkey("ctrl","enter")




```
